1. Veri kümesinde yer alan değişkenlerinin tiplerini ve her bir değişkenin eksik (null) değer oranını bulun.

```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
states_all = pd.read_csv("Odevler/Bolum4/states_all.csv")
states_all.head(10)
```

```python
#Veri kümesinde yer alan değişkenlerinin tipleri: 

states_all.info()
```

```python
#Her bir değişkenin eksik (null) değer oranını bulun.
missing_values= states_all[(states_all["PRIMARY_KEY"].str.strip() == "--") | (states_all["STATE"].str.strip() == "--")] 

display(missing_values.head())
print('missing_values: ',  len(missing_values))
```

BURDA FORMULDE INTEGER OLAN VERILER ICIN EKSIK DEGERI BULMUYOR


missing_values= states_all[(states_all["AVG_MATH_8_SCORE"].int.strip() == "--") | (states_all["AVG_MATH_4_SCORE"].int.strip() == "--")] 

display(missing_values.head())
print('missing_values: ',  len(missing_values))

```python
#BURDA DA BIR SONUC GELMEDI


for sutun_adi in ["AVG_READING_8_SCORE", "AVG_MATH_8_SCORE", "AVG_MATH_4_SCORE", "AVG_READING_4_SCORE", "STATE", "YEAR"]:
    print("{} sütunu için problemli değerler : ".format(sutun_adi))
    for deger in states_all["AVG_READING_8_SCORE"]:
        try:
            float(deger)
        except:
            print(deger)
```

```python
# Eksik Degerler Nasil tespit edilir? 

#Pandas, bir veri çerçevesindeki her öğe için bir öğenin değeri yoksa (None) veya False ifadesini alırsa, True döndüren 
#.isnull () adlı bir metoda sahiptir. Ancak, bu metodu veri çerçevemizden çağırdığımızda, şunu elde ederiz:


states_all.isnull().head()
```

```python
#Benzersiz değerleri incelemek için veri çerçevesinin sütunlarında .unique () yöntemini çağırabiliriz. 
#Örneğin, verilerdeki her bir sütuna ait tüm benzersiz değerleri almak için:

for sutun_adi in states_all.columns:
    print("{} sütunundaki benzersiz değerler : {}".format(sutun_adi, states_all[sutun_adi].unique()))
```

```python
states_all.isnull().sum()*100/states_all.count()
```

```python
states_all.STATE.value_counts()
```

```python
states_all.PRIMARY_KEY.value_counts()
```

```python
#Kayıp değerleri sayarken etraflarında bazı boşluklar olduğu için görünmeyen bazı boş dizge değerlerimiz var gibi görünüyor. 
#Onları np.nan a dönüştürelim ve değişikliklerimizin etkili olup olmadığını kontrol edelim:

states_all.STATE = states_all.STATE.apply(str.strip).replace("", np.nan)

states_all.STATE.unique()
```

2. Verimizde yıl (year) sütunu olduğunu farketmişsinizdir. Şimdilik yıl verisini unutun ve her bir gözlemin aynı yıl içerisinde yapıldığını farz edin. Her bir değişken için eksik değerleri nasıl doldurabileceğinizi düşünün. Eksik değerleri bir değerle doldurmak hangi değişkenler için anlamlı, hangileri için anlamsızdır?

```python
states_all["ENROLL"].fillna(states_all["ENROLL"].mean(), inplace=True)
states_all["AVG_MATH_4_SCORE"].fillna(states_all["AVG_MATH_4_SCORE"].mean(), inplace=True)
states_all["YEAR"].fillna(states_all["YEAR"].mean(), inplace=True)
states_all["TOTAL_REVENUE"].fillna(states_all["TOTAL_REVENUE"].mean(), inplace=True)
states_all["FEDERAL_REVENUE"].fillna(states_all["FEDERAL_REVENUE"].mean(), inplace=True)
states_all["STATE_REVENUE"].fillna(states_all["STATE_REVENUE"].mean(), inplace=True)
states_all["LOCAL_REVENUE"].fillna(states_all["LOCAL_REVENUE"].mean(), inplace=True)
states_all["TOTAL_EXPENDITURE"].fillna(states_all["TOTAL_EXPENDITURE"].mean(), inplace=True)
states_all["INSTRUCTION_EXPENDITURE"].fillna(states_all["INSTRUCTION_EXPENDITURE"].mean(), inplace=True)
states_all["SUPPORT_SERVICES_EXPENDITURE"].fillna(states_all["SUPPORT_SERVICES_EXPENDITURE"].mean(), inplace=True)
states_all["OTHER_EXPENDITURE"].fillna(states_all["OTHER_EXPENDITURE"].mean(), inplace=True)
states_all["CAPITAL_OUTLAY_EXPENDITURE"].fillna(states_all["CAPITAL_OUTLAY_EXPENDITURE"].mean(), inplace=True)
states_all["GRADES_PK_G"].fillna(states_all["GRADES_PK_G"].mean(), inplace=True)
states_all["GRADES_KG_G"].fillna(states_all["GRADES_KG_G"].mean(), inplace=True)
states_all["GRADES_4_G"].fillna(states_all["GRADES_4_G"].mean(), inplace=True)
states_all["GRADES_8_G"].fillna(states_all["GRADES_8_G"].mean(), inplace=True)
states_all["GRADES_12_G"].fillna(states_all["GRADES_12_G"].mean(), inplace=True)
states_all["GRADES_1_8_G"].fillna(states_all["GRADES_1_8_G"].mean(), inplace=True)
states_all["GRADES_9_12_G"].fillna(states_all["GRADES_9_12_G"].mean(), inplace=True)
states_all["GRADES_ALL_G"].fillna(states_all["GRADES_ALL_G"].mean(), inplace=True)
states_all["YEAR"].fillna(states_all["YEAR"].mean(), inplace=True)
states_all["AVG_MATH_8_SCORE"].fillna(states_all["AVG_MATH_8_SCORE"].mean(), inplace=True)
states_all["AVG_READING_4_SCORE"].fillna(states_all["AVG_READING_4_SCORE"].mean(), inplace=True)
states_all["AVG_READING_8_SCORE"].fillna(states_all["AVG_READING_8_SCORE"].mean(), inplace=True)
```

#Burda degerler string oldugu icin bu formuller calismadi. Nasil eksik degerleri doldurabilirim?

states_all["STATE"].fillna(states_all["STATE"].mean(), inplace=True)
